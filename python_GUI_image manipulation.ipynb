```python
import PySimpleGUI as sg
from PIL import Image,ImageTk,ImageFilter,ImageEnhance

title="Edit : Browse and Select Image"
layout=[[sg.FileBrowse(key="file_name", size=(10,2))],
        [sg.Submit("OK")],
        [sg.Button("Blur",key="blur"),
         sg.Button("Horizontal_flip",key="flip"),
         sg.Button("Vertical_flip",key="flip2"),
         sg.Button("Minimise_size",key="resize"),
         sg.Button("Black & white",key="color")],
        [sg.Text("degree of rotation:",size=(10,2)),sg.Input(key="degree")],
        [sg.Button("Rotate",key ="rotate")],
        [sg.Image(key="showimage")],
        [sg.Button("Save_Image",key="saveimage"),
     #    sg.Button("Save_Image_pdf", key="saveimage"),
         sg.Cancel("exit")]]

window=sg.Window(title,layout)

while True:
    event,values = window.read()
    print("event " , event)
    print("values " , values)
    if event == sg.WINDOW_CLOSED or event=="Exit":
        break
    elif event == "OK":
        img = Image.open(values["file_name"])
        window["showimage"].update(data = ImageTk.PhotoImage(img))
    elif event == "blur":
        img = img.filter(ImageFilter.GaussianBlur(5))
        window["showimage"].update(data = ImageTk.PhotoImage(img))
    elif event == "flip":
        img = img.transpose(Image.FLIP_LEFT_RIGHT)
        window["showimage"].update(data = ImageTk.PhotoImage(img))
    elif event == "flip2":
        img = img.transpose(Image.FLIP_TOP_BOTTOM)
        window["showimage"].update(data = ImageTk.PhotoImage(img))
    elif event == "resize":
        img = img.resize((round(img.size[0]*0.4),round(img.size[1]*0.4)))
        window["showimage"].update(data = ImageTk.PhotoImage(img))
        img.save("A:\\user\\Image_New_min.png")
    elif event == "color":
        img = ImageEnhance.Color(img)
        img.enhance(0.0).show()
        window["showimage"].update(data = ImageTk.PhotoImage(img))
        img.save("A:\\user\\Image_New_bnw.png")
    elif event == "rotate":
        degree = values["degree"]
        img = img.rotate(float(degree))
        window["showimage"].update(data = ImageTk.PhotoImage(img))
    elif event == "saveimage":
        img.save("A:\\user\\Image_New.png")
    else:
        
        break
    
window.close()
```
